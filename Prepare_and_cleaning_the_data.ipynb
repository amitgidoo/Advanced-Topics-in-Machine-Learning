```python
import pandas as pd
import numpy as np
```

```python
# Load the Excel file into a DataFrame
file_name = "final_merge_tel_aviv_data.xlsx"
df = pd.read_excel(file_name)
```

```python
df.info()
```

```python
# Check for duplicate rows
duplicate_rows = df[df.duplicated()]

if duplicate_rows.shape[0] == 0:
    print("No duplicate rows found.")
else:
    print("Duplicate rows found:")
    print(duplicate_rows)
```

```python
# נניח שערכים לא ריאליים הם מחירים שהם 0 או גדולים מאוד, ושטחים שהם 0 או גדולים מאוד
# ננקה את הנתונים מערכים אלו
df_cleaned = df[(df['Price'] > 1000) & (df['Price'] < 50000)]  # מחיר סביר
df_cleaned = df[(df['Size'] > 10) & (df['Size'] < 500)]  # שטח סביר
```

```python
df_cleaned.info()
```

```python
data_filtered_cleaned = df_cleaned.dropna()
```

```python
data_filtered_cleaned.info()
```

```python
# Task 1: Extract year from the Date column and add it to a new column called Year
data_filtered_cleaned['Year'] = data_filtered_cleaned['Date Post'].dt.year

# Task 2: Format the date in Month-Day format and add it to a new column called Month_Day_Date
data_filtered_cleaned['Month'] = data_filtered_cleaned['Date Post'].dt.strftime('%m')
```

```python
# Reset the index of the sorted DataFrame
data_filtered_cleaned.reset_index(drop=True, inplace=True)
```

```python
data_filtered_cleaned
```

```python
data_filtered_cleaned.to_excel('clean_final_data.xlsx', index=False)
```
